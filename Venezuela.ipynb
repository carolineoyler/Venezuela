```
venezuela = pd.read_excel('Venezuela Data .xlsx')
venezuela.head()
```

```
venezuela.columns = ['year', 'CPI new measurement', 'CPI_old', 'income inequality', 'political_regime', 'GDP per capita growth', 'economic_freedom', 'human_rights', 'aid_received_pc', 'GDP_pc', 'population_poverty', 'exports as percentage of GDP', 'unemployment_rate', 'Foreign direct investment as percentage of GDP', 'Oil rents as percentage of GDP', 'Fuel exports as percentage of merchandise exports', 'human rights violations']
```

```
venezuela.head(22)
```

```
import matplotlib.pyplot as plt
```

```
plt.bar(venezuela.year,venezuela.political_regime)
plt.title('Democracy Scores 1995-2015')
plt.xlabel('Years')
plt.ylabel('Regime Score')
plt.show()
#This visualisation is based on the Polity IV data: 
#The Polity IV score captures the type of political regime for each country on a range from -10 (full autocracy) 
#to +10 (full democracy). Regimes that fall into the middle of this spectrum are called anocracies.
```

```
plt.plot(venezuela['year'], venezuela['CPI_old'])
plt.title('Corruption Index')
plt.xlabel('Year')
plt.ylabel('Corruption Perception')
plt.show()
#Transparency International's Corruption Perception Index. 
#Scores are on a scale of 0-10, where 0 means that a country is perceived as highly corrupt.
```

```
plt.plot(venezuela['year'], venezuela['human_rights'])
plt.title('Human Rights Score')
plt.xlabel('Year')
plt.ylabel('Human Rights Score')
plt.show()
#These Human Rights Scores indicate the degree to which governments protect and respect human rights. 
#The values range from around 3.8 to around 5.4 (the higher the better).
```

```
plt.plot(venezuela['year'], venezuela['economic_freedom'])
plt.title('Economic Freedom Score')
plt.xlabel('Year')
plt.ylabel('Economic Freedom')
plt.show()
#Economic Freedom of the World is calculated by the Fraser Institute, 
#and measures the degree to which individuals are free to choose, trade, and cooperate with
#others, and compete as they see fit. Scores are on a scale of 0-10, where 10 represents maximum economic freedom.
```

```

plt.plot(venezuela['year'], venezuela['GDP_pc'])
plt.title('Gross Domestic Product Per Capita')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.show()
```

```

plt.plot(venezuela['year'], venezuela['population_poverty'])
plt.title('Percentage of the Population Living Below the National Poverty Line')
plt.xlabel('Year')
plt.ylabel('Poverty')
plt.show()
```

```
plt.plot(venezuela['year'], venezuela['unemployment_rate'])
plt.title('Percentage of the Population Unemployed')
plt.xlabel('Year')
plt.ylabel('Unemployment')
plt.show()
```

```
# create a figure and axis
fig, ax = plt.subplots()

# scatter the variables 
ax.scatter(venezuela['CPI_old'], venezuela['economic_freedom'])

# label the graph 
ax.set_title('Corruption and Economic Freedom')
ax.set_xlabel('Corruption')
ax.set_ylabel('Economic Freedom')

#economic freedom, scale 0-10 with 10 being most freedom 
#corruption perception scale of 1-10 with 10 being least corruption 
```
